```python
# Remove rows where all BATTING and BOWLING Mat columns are null or empty
bat_cols = ["BATTING_Tests_Mat", "BATTING_T20Is_Mat", "BATTING_ODIs_Mat"]
bowl_cols = ["BOWLING_Tests_Mat", "BOWLING_T20Is_Mat", "BOWLING_ODIs_Mat"]
def is_row_valid(row):
    return any(pd.notnull(row[c]) and str(row[c]).strip() != '' for c in bat_cols + bowl_cols)
df_cleaned = df[df.apply(is_row_valid, axis=1)]
df_cleaned
```
